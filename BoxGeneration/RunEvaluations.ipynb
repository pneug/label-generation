```python
# a method that makes a given plot look more professional
def prettify_plot(ax):
    # let the diagram look more professional
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # add horizontal lines
    ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=0.3) # 0.25
    
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
```

```python
import os

def save_plot(fig, save_path):
    if save_path is not None:
        # check if the folder exists
        folder = os.path.dirname(save_path)
        if not os.path.exists(folder):
            os.makedirs(folder)
        fig.savefig(save_path, bbox_inches='tight', pad_inches=0, dpi=200)
```

```python
import numpy as np
# plot the pareto frontier to see the tradeoff between acceptance rate and number of images
# y axis is acceptance rate, x axis is number of images
# plot the results

def is_pareto_efficient(costs):
    """
    Returns a boolean mask indicating which points are Pareto efficient
    """
    is_efficient = np.ones(costs.shape[0], dtype=bool)
    for i, c in enumerate(costs):
        if is_efficient[i]:
            # remove points for which the same accuracy is achieved with less images
            is_efficient[is_efficient] = np.any(costs[is_efficient] > c, axis=1)  # Remove dominated points
            is_efficient[i] = True  # Keep self
    return is_efficient

def plot_paretor_front(settings, results, results_cam5, x_label):

    fig, ax = plt.subplots()
    prettify_plot(ax)
    x = []
    y = []
    for min_shared_features in settings["min_shared_features"]:
        good_images = results[(min_shared_features, "median")]["Good"]
        good_images_cam5 = results_cam5[(min_shared_features, "median", "cam_5_only")]["Good"]
        all_images = sum(results[(min_shared_features, "median")].values())
        all_images_cam5 = sum(results_cam5[(min_shared_features, "median", "cam_5_only")].values())
        if all_images == 0:
            y.append(0)
        else:
            y.append(good_images / all_images)
        if all_images_cam5 == 0:
            y.append(0)
        else:
            y.append(good_images_cam5 / all_images_cam5)
        num_good_images = results[(min_shared_features, "median")]["Good"]
        num_good_images_cam5 = results_cam5[(min_shared_features, "median", "cam_5_only")]["Good"]
        x.append(num_good_images)
        x.append(num_good_images_cam5)
        # x.append(all_images)
    # ax.scatter(x, y)
    ax.set_ylabel('Accuracy', fontsize=16)
    ax.set_xlabel(x_label, fontsize=16) # 'Number of good images'
    # ax.set_title('Acceptance rate vs number of good images')
    ax.set_xticks(x)
    # set the x ticks labels in a way that they dont overlap
    
    ax.set_xticklabels([str(i) if i not in [14, 19] else "" for i in x])


    # ax.set_xticklabels(x)
    # ax.set_xticklabels(settings["min_shared_features"])
    # ax.legend()

    # mark for each point the min_shared_features
    # for i, txt in enumerate(settings["min_shared_features"]):
    #     ax.annotate(txt, (x[i], y[i]))

    # combine labels that are plotted on top of each other
    annotations = {}
    for i, msf in enumerate(settings["min_shared_features"]):
        for j in range(2):
            if j == 0:
                pos = (x[2 * i] + 0.00, y[2 * i] + 0.01)
            else:
                pos = (x[2 * i + 1] + 0.00, y[2 * i + 1] - 0.03)

            if pos in annotations:
                if annotations[pos] == msf:
                    continue
                if msf < int(annotations[pos]):
                    annotations[pos] = str(msf)#  + "," + annotations[pos]
                # else:
                    # annotations[pos] += "," + str(msf)
            else:
                annotations[pos] = str(msf)

    for pos, txt in annotations.items():
        ax.annotate(txt, pos, fontsize=16)



    # # mark for each point the min_shared_features
    # for i, txt in enumerate(settings["min_shared_features"]):
    #     ax.annotate(txt, (x[2 * i] + 0.02, y[2 * i] + 0.01))
    #     ax.annotate(txt, (x[2 * i + 1] + 0.02, y[2 * i + 1] - 0.03))
        # remember that these labels have been plotted to avoid overlap with the other labels



    x = np.array(x)
    y = np.array(y)

    # Combine x and y coordinates into a single array
    points = np.column_stack((x, y))

    # Find Pareto efficient points
    is_efficient = is_pareto_efficient(points)

    # Extract Pareto front coordinates
    pareto_x = x[is_efficient]
    pareto_y = y[is_efficient]

    # sort the pareto front
    pareto_x, pareto_y = zip(*sorted(zip(pareto_x, pareto_y)))

    # Plot original points
    # plt.scatter(x, y, label='All Points')

    # color every second point differently
    for i in range(len(x)):
        if i % 2 == 0 and not i == 14:
            ax.scatter(x[i], y[i], color="lightgreen")
        else:
            ax.scatter(x[i], y[i], color="blue")

    # show the legend
    ax.legend(["All cameras", "Camera 5 only"], fontsize=16)

    # Plot Pareto front
    plt.plot(pareto_x, pareto_y, label='Pareto Front', color='r')


    # draw in the pareto frontier
    # highest_point = (0, 0)
    # for i in range(len(x)):
    #     if y[i] > highest_point[1]:
    #         highest_point = (x[i], y[i])

    # # draw a line from the highest point to the next point on the frontier
    # for i in range(len(x)):
    #     if x[i] <= highest_point[0]:
    #         continue
        

    # for i in range(len(x)):
    #     if i == 0:
    #         continue
    #     if y[i] > y[i-1]:
    #         ax.plot([x[i-1], x[i]], [y[i-1], y[i]], color="red")

    fig.tight_layout()

    plt.show()

    save_plot(fig, "../outputs/plots/pareto_frontier.pdf")

plot_paretor_front(settings, results_labels, results_cam5_labels, "Number of good Labels")
# plot_paretor_front(settings, results_images, results_cam5_images, "Number of good Images")
```

```python
def split_into_train_val_test(label_path, train_split, val_split, test_split):
    with open(label_path) as f:
        gt_labels = json.load(f)
    
    # get the image ids
    image_ids = [img["id"] for img in gt_labels["images"]]
    num_images = len(image_ids)
    num_train = int(num_images * train_split)
    num_val = int(num_images * val_split)
    num_test = int(num_images * test_split)
    train_split = image_ids[:num_train]
    val_split = image_ids[num_train:num_train+num_val]
    test_split = image_ids[num_train+num_val:]
    print("Num train: ", len(train_split))
    print("Num val: ", len(val_split))
    print("Num test: ", len(test_split))
    # save the splits to a json file
    train_split_dict = {"images": [], "annotations": []}
    val_split_dict = {"images": [], "annotations": []}
    test_split_dict = {"images": [], "annotations": []}
    for img in gt_labels["images"]:
        if img["id"] in train_split:
            train_split_dict["images"].append(img)
        elif img["id"] in val_split:
            val_split_dict["images"].append(img)
        elif img["id"] in test_split:
            test_split_dict["images"].append(img)
        else:
            print("Image id not found in any split!")
    for ann in gt_labels["annotations"]:
        if ann["image_id"] in train_split:
            train_split_dict["annotations"].append(ann)
        elif ann["image_id"] in val_split:
            val_split_dict["annotations"].append(ann)
        elif ann["image_id"] in test_split:
            test_split_dict["annotations"].append(ann)
        else:
            print("Annotation image id not found in any split!")

    # add the other data required for coco format
    train_split_dict["info"] = gt_labels["info"]
    train_split_dict["licenses"] = gt_labels["licenses"]
    train_split_dict["categories"] = gt_labels["categories"]
    val_split_dict["info"] = gt_labels["info"]
    val_split_dict["licenses"] = gt_labels["licenses"]
    val_split_dict["categories"] = gt_labels["categories"]
    test_split_dict["info"] = gt_labels["info"]
    test_split_dict["licenses"] = gt_labels["licenses"]
    test_split_dict["categories"] = gt_labels["categories"]
    
    # save the splits to a json file
    train_name = label_path[:-5] + "_train.json"
    with open(train_name, "w") as f:
        json.dump(train_split_dict, f)
    val_name = label_path[:-5] + "_val.json"
    with open(val_name, "w") as f:
        json.dump(val_split_dict, f)
    test_name = label_path[:-5] + "_test.json"
    with open(test_name, "w") as f:
        json.dump(test_split_dict, f)

    return train_name, val_name, test_name

label_path = "../data/Waste_Bin_Detection_Dataset/sunny_2021_03_23_14_33_cam5_filtered (training dataset images and ground truths)/"
label_path += "sunny_reduced_static.json"
split_into_train_val_test(label_path, 0.6, 0.2, 0.2)
```
