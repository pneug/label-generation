```python


def dms2dd(degrees, minutes, seconds, direction):
    dd = float(degrees) + float(minutes)/60 + float(seconds)/(60*60)
    if direction == 'E' or direction == 'S':
        dd *= -1
    return dd

def get_gps(file_name):
    exif_data = piexif.load(file_name)
    if "GPS" in exif_data:
        gps = exif_data['GPS']
        print("gps", gps)
        dd_lat = dms2dd(gps[2][0][0] / gps[2][0][1], gps[2][1][0] / gps[2][1][1], gps[2][2][0] / gps[2][2][1], gps[1])
        dd_lon = dms2dd(gps[4][0][0] / gps[4][0][1], gps[4][1][0] / gps[4][1][1], gps[4][2][0] / gps[4][2][1], gps[3])
        dd = (dd_lat, dd_lon)
        return dd

```
