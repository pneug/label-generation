```python
# Trained on reduced original dataset" (HL)
# |   AP   |  AP50  |  AP75  |  APs   |  APm   |  APl   |
# | 42.227 | 58.580 | 51.519 | 20.045 | 49.418 | 77.425 | dataset_test_original_reduced
# | 23.288 | 35.369 | 27.442 | 12.593 | 30.684 | 50.503 | dataset_snowy_day
# | 30.954 | 42.883 | 39.340 | 6.343  | 33.531 | 49.271 | dataset_snowy_night
# | 26.906 | 37.472 | 34.993 | 10.497 | 28.353 | 67.791 | dataset_rb
# Trained on reduced original ds + snowy train set (HL + SG)
# | 40.526 | 55.548 | 47.742 | 18.466 | 49.051 | 72.051 | dataset_test_original_reduced
# | 22.646 | 32.486 | 29.322 | 14.175 | 28.426 | 45.474 | dataset_snowy_day
# | 32.458 | 44.207 | 39.656 | 9.361  | 37.541 | 42.727 | dataset_snowy_night
# | 30.169 | 43.084 | 40.356 | 11.836 | 30.764 | 70.756 | dataset_rb
# Trained on snowy train set (SG)
# | 27.400 | 43.050 | 34.106 | 12.741 | 32.122 | 55.029 | dataset_test_original_reduced
# | 13.037 | 20.674 | 16.070 | 8.540  | 17.123 | 28.456 | dataset_snowy_day
# | 16.335 | 25.245 | 22.863 | 9.003  | 19.103 | 22.028 | dataset_snowy_night
# | 19.437 | 32.808 | 26.297 | 9.343  | 24.582 | 49.679 | dataset_rb
# Trained on reduced original ds+snowy12_clean train set (HL + SC)
# | 40.249 | 55.174 | 46.785 | 18.387 | 45.818 | 75.338 | dataset_test_original_reduced
# | 22.504 | 33.711 | 27.502 | 12.444 | 26.322 | 50.499 | dataset_snowy_day
# | 30.741 | 42.843 | 37.571 | 8.163  | 36.492 | 40.848 | dataset_snowy_night
# | 28.301 | 39.012 | 35.747 | 10.758 | 26.531 | 72.088 | dataset_rb"

# read in the Results.txt file which looks like the comment above
with open('Results.txt', 'r') as file:
    data = file.readlines()

# create a dict of dicts with the data
results = {}
training_sets = ["HL", "HL+SG", "SG", "HL+SC"]
test_sets = ["original", "snowy_day", "snowy_night", "rb"]
for i, line in enumerate(data):
    if i % 5 == 0:
        training_set = training_sets[i // 5]
        training_results = {}
    else:
        test_set = test_sets[(i - 1) % 5]
        line = line.split('|')
        training_results[test_set] = {'AP': line[1].strip(),
                                    'AP50': line[2].strip(),
                                    'AP75': line[3].strip(),
                                    'APs': line[4].strip(),
                                    'APm': line[5].strip(),
                                    'APl': line[6].strip()}
        if (i - 1) % 5 == 3:
            results[training_set] = training_results

print(results)

# round all results to only 1 decimal place
for training_set in results:
    for test_set in results[training_set]:
        for metric in results[training_set][test_set]:
            results[training_set][test_set][metric] = round(float(results[training_set][test_set][metric]), 1)

# create a latex table
refs = ["waste-bin-test", "snowy-test", "snowy-test", "roadbotics-test"]
test_set_latex_names = ["cloudy", "snowy day", "snowy night", "RoadBotics"]
hline = "\\hline \n"
code =  "\\begin{tabularx}{\\textwidth}{p{3.5cm}XXXXX} \n"
code += " &  & HL & SG & HL + SG & HL + SC \\\\ \n"
for i, test_set in enumerate(test_sets):
    code += hline
    code += f"{test_set_latex_names[i]} [~\\ref{{{refs[i]}}}] & AP \\newline AP50 "
    # & {} \\newline 58.6 & 27.4 \\newline 43.1 & 40.5 \\newline 55.5 & 40.2 \\newline 55.17 \\\\
    for training_set in training_sets:
        code += f"& {results[training_set][test_set]['AP']} \\newline {results[training_set][test_set]['AP50']} "
    code += "\\\\ \n"

code += "\\end{tabularx}"
print(code)
```
